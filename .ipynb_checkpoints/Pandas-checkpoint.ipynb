```python
# Q1 -> find all the final winners in ipl dataframe
ipl
```

```python
#we take it step wise
#first find all the final matches
final = ipl['MatchNumber'] == 'Final'
final
```

```python
#Above we have a boolean of which matches are final
#Below we have dataframe with Final as the value of MatchNumber column
#We will save this dataframe in a new dataframe
new_df = ipl[final]
new_df
```

```python
#Now we have to take out the Winning Team from new dataframe and store it in new data frame
winning_team = new_df[['Season','WinningTeam']]
winning_team
```

```python
#We can also do it in a single line
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
#This gave us winning teams
#Q1 END
```

```python
# Q2 -> how many super over finishes have occured
#here shape will give us the dim of dataframe but we need only rows so we make it shape[0]
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# Q3 -> how many matches has csk won in kolkata
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
#IMPORTANT --> we used '&' because in this we got boolean series and for that we need binary operators and not logical operators
```

```python
# Q4 -> toss winner is match winner in percentage
ipl['TossWinner'] == ipl['WinningTeam']
#This compares both the columns and checks if both toss winner and winning team are same then true otherwise false
```

```python
#all the matches where this is true
ipl[ipl['TossWinner'] == ipl['WinningTeam']]
```

```python
#total number of such matches
ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]
```

```python
(     ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]      /     ipl.shape[0])     *100
#this is simple favourable/total times 100 for percentage
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5

points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
