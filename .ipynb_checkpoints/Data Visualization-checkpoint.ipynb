```python
#fix
plt.bar(colors,children)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')

plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')

plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

# Seaborn

### Why Seaborn?

- provides a layer of abstraction hence simpler to use than matplotlib
- better aesthetics
- more graphs included

### Seaborn Roadmap

Types of Functions

- Figure Level
- Axis Level

Main Classification

- Relational Plot -> Scatter Plot, Line Plot
- Distribution Plot -> Histogram, KDE Plot, Rug Plot
- Categorical Plot -> Bar Plot, Count Plot, Box Plot, Violin, Swam Plot
- Regression Plot -> 
- Matrix Plot -> Heat Map, Cluster Map
- Multiplots -> Joint Plot, Pair Plot

https://seaborn.pydata.org/api.html

### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
tips
```

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
# relplot -> figure level
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time',size='size')

#using relplot we didn't need to write a lot of things as most of them are done by sns itself
```

```python
# line plot
gap = px.data.gapminder()
gap.sample(5)
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relpplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df.sample(5)
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot

#sns.relplot(data=tips, x='total_bill', y='tip', kind='line', hue='sex')
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
#cannot use row and col_wrap together
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
#sns.histplot(data=tips, x='total_bill')
sns.displot(data=tips, x='total_bill', kind='hist',hue='sex')
```

```python
# element -> step just for better visualization
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist',bins=2)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.(basically days ko x axis par daal diya lekin technically yeh ek count plot hai)
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=tips, x='day', kind='hist')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

### KDEplot

```python
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
#y axis is probability

#sns.kdeplot(data=tips,x='total_bill')
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

### Rugplot

```python
# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram basically 2d histogram
# A bivariate histogram bins the data within rectangles that tile the plot ie both axes and then shows the frequency with fill color with darker colour means higher frequecy

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
# lines closer means higher frequency
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 3. Matrix Plot

- Heatmap
- Clustermap

Heatmap -> Plot rectangular data as a color-encoded matrix

```python
# Heatmap
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

temp_df.sample(5)
```

```python
# axes level function
#there is no figure level function to pass a heat map
sns.heatmap(temp_df)
```

```python
# annot gives number as well
#linewidth gives gap between blocks
#cmap is colour map that it colours of grids
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

sns.heatmap(temp_df,annot=True,linewidth=0.5)
```

Clustermap ->Plot a matrix dataset as a hierarchically-clustered heatmap.

```python
# Clustermap
# This function requires scipy to be available.

iris = px.data.iris()
iris.head(3)
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

### Boxplot(Important)

- A boxplot is a standardized way of displaying the distribution of data based on a five number summary
- (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”).
- It can tell you about your outliers and what their values are.
- Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box plot - axes level
# Using catplot - figure level
#sns.catplot(data=tips,x='day',y='total_bill',kind='box')
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

### Violinplot = (Boxplot + KDEplot)
Violin Plot shows distribution using KDE and shows value using Boxplot

```python
# violinplot
#sns.violinplot(data=tips,x='day',y='total_bill')
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# barplot
# some issue with errorbar
sns.barplot(data=tips, x='sex', y='total_bill',errorbar=None)
```

```python
# point plot is same as bar plot minus the bars and all error bars conneceted
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker')
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker')
```

```python
# countplot
# is a special case of bar plot when you want to show the number of observations in each category rather than computing a statistic for a second variable 
sns.countplot(data=tips,x='sex',hue='day')
```

```python
# faceting using catplot
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

### A second way to plot Facet plots -> FacetGrid

```python
# M1 -> sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')

# M2 ->
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
#pairplot plots all the graphs made with all possible pair of column
sns.pairplot(iris)
```

```python
# pair grid
g = sns.PairGrid(data=iris)

g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

#This method gives freedom to choose the type of graph we need

#OR
# g.map(sns.scatterplot) but less freedom
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.lineplot)
g.map_lower(sns.scatterplot)
```

```python
# vars lets you choose the columns you want to make graphs of
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid Vs Jointplot

```python
#jointplot
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
#scatterplot and histogram
```

```python
#JointGrid
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
