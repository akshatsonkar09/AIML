```python
#NUMPY ARRAY VS PYTHON LIST
#numpy aaya cause python was slow
#numpy is just C wrapped python
```

```python
#Speed

#list
a = [i for i in range(10000000)]   #[0, 1, 2, 3, 4, 5, ..., 9999997, 9999998, 9999999]
b = [i for i in range(10000000,20000000)]   #[10000000, 10000001, ...,  19999998, 19999999]

c = []
```

```python
 #this is a module which lets us deal with real time 
import time
```

```python
start = time.time() #this is like a stopwatch which saves the time at this point in start variable

for i in range(len(a)):
  c.append(a[i] + b[i])
print(time.time()-start)  #this subtracts the start time from current time
```

```python
# numpy
a = np.arange(10000000)
b = np.arange(10000000,20000000)

start = time.time()
c = a + b
print(time.time()-start)
```

```python
0.9769461154937744/0.02992415428161621
```

```python
#32 times faster in this case
#reason for this is numpy uses C type array ie its size is fixed, stores homogeneous datatypes and it stores data in memory and not its address like in python list
```

```python
#BROADCASTING

#The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.
#The smaller array is “broadcast” across the larger array so that they have compatible shapes.
#In simple words it is how the numpy broadcasts that is enlarges the smaller array so that the arithmetic operation is possible. 
```

```python
# diff shape
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a+b)
```

```python
#Broadcasting rules
#1) Make a number of dimensions same Eg-> (3,3):(3) -> (3,3):(1,3)
#2) Make the value/size of dimension same Eg-> (3,3):(1,3) -> (3,3):(3,3)
# inrule no.1 1 is added to the front
#in rule no.2 1 is stretched until it is same as the needed size and if there isno one then cannot be broadcasted
```

![Broadcasting Image](https://jakevdp.github.io/PythonDataScienceHandbook/figures/02.05-broadcasting.png)

```python
# Mean Squared Error

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
```

```python
actual
```

```python
predicted
```

```python
def mse(actual,predicted):
  return np.mean((actual - predicted)**2)

#formula was (x -xnot)^2/N

mse(actual,predicted)
```
