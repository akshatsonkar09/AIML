```python
# agg method allows us to do multilple different functions on different columns at once

# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list

#genres.agg(['min','max','mean'])   for some reason not working
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
df = pd.DataFrame(columns=movies.columns)
for group,data in genres:
  temp = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
  df = pd.concat([df, temp])
df
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
# find number of movies starting with A for each group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
    
genres.apply(foo)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)  
```

```python
# pd.concat concatenate vertically
regs = pd.concat([nov,dec],ignore_index=True)
regs

#M2 -> nov.append(dec,ignore_index=True) 
```

```python
#Join	Meaning
#inner	common rows only
#left	keep all left rows
#right	keep all right rows
#outer	keep everything
```

```python
# inner join print common data in column 
#See how there is no student id beyond 25
#left_table.merge(right_table,how='type of join', on='common column')
students.merge(regs,how='inner',on='student_id')
```

```python
# left join means left will print data even if not common
#course id 11 and 12 no common still printed
courses.merge(regs,how='left',on='course_id')
```

```python
# right join
students.merge(regs,how='right',on='student_id')
```

```python
# outer join
students.merge(regs,how='outer',on='student_id')
```

```python
# Q-> Print the registration table
# cols -> name -> course -> price
regs.merge(students,on='student_id').merge(courses,on='course_id')[['name','course_name','price']]
```

```python
# Q-> find students who enrolled in both the months
common_student_id = np.intersect1d(nov['student_id'],dec['student_id'])
common_student_id
```

```python
students[students['student_id'].isin(common_student_id)]
```

```python
# Q-> Print student name -> partner name for all enrolled students
# self join
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```
